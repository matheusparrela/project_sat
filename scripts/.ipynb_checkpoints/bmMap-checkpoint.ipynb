```python
import json
import time
import folium
```

```python
# Mudando o tipo de gráfico
m = folium.Map(location=[-16.2063900, -44.4333300], # Coordenadas de Brasília de Minas
               zoom_start=10, tiles= 'Stamen Terrain')
```

Dados sobre o limite territórial da cidade de Brasília de Minas e região

```python
def read_json():
    try:
        dict = None
        # Opening JSON file
        file = open('data\data.json',)
        # returns JSON object as a dictionary
        dict = json.load(file)
        # Closing file
        file.close()
        print(dict)
        return dict
    except:
        print('Falha')
        time.sleep(2)
        read_json()
  
dict = read_json()
```

```python
print(len(dict['features'][0]['geometry']['coordinates'][0]))
print(dict['features'][0]['geometry']['coordinates'])
```

```python
for j in range(0,14-1):
    
    coords = []
    temp = []
    for i in range (0 , len(dict['features'][j]['geometry']['coordinates'][0])):

      temp.append(dict['features'][j]['geometry']['coordinates'][0][i])
      temp[i][0], temp[i][1] = temp[i][1], temp[i][0] 
      coords.append(tuple(temp[i]))
    print(coords)
    folium.PolyLine(coords, tooltip="Trail").add_to(m)
    
```
