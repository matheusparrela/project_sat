Dados sobre o limite territórial da cidade de Brasília de Minas

```python
def read_json():
    try:
        dict = None
        # Opening JSON file
        file = open('data.json', )
        # returns JSON object as a dictionary
        dict = json.load(file)
        # Closing file
        file.close()
        return dict
    except:
        print('Falha')
        time.sleep(2)
        read_json()
  
dict = read_json()
```

```python
# Mudando o tipo de gráfico
m = folium.Map(location=[-16.2063900, -44.4333300],
               zoom_start=11, tiles= 'Stamen Terrain')

coords = []
temp = []
for i in range (0 , len(dict['features']["name" == "Montes Claros"]['geometry']['coordinates'][0])):
  
  temp.append(dict['features']["name" == "Montes Claros"]['geometry']['coordinates'][0][i])
  temp[i][0], temp[i][1] = temp[i][1], temp[i][0] 
  coords.append(tuple(temp[i]))
  
folium.PolyLine(coords, tooltip="Trail").add_to(m)
print(coords)
```
